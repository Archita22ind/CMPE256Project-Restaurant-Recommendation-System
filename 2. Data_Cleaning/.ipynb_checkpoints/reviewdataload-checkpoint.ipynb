#### Loading user data from Yelp Dataset

```python
def loadReviewData():
    global reviews
    with open(path, encoding='utf-8') as fin:
        i = 0 
        for line in fin:
            line_contents = json.loads(line)
            reviews.append(line_contents)
    reviews = pd.DataFrame(reviews)
    reviews.shape
    return
```

#### Writing Processed Reviews data to CSV format

```python
def filetoCSV():
    reviews.to_csv("/content/drive/MyDrive/CMPE256/Data/CSVfiles/California_reviews_c.csv")
```

#### Getting California Specific Reviews

```python
def getCAReviews():
    getUniqueCABusinessIds = business.business_id.unique()
    reviewIds = []
    global reviews
    for i in range(reviews.shape[0]):
        if reviews.business_id[i] in getUniqueCABusinessIds:
            reviewIds.append(i)

    reviews = reviews.iloc[reviewIds,:]
    reviews.index = range(reviews.shape[0])

    yearList = []
    for x in reviews.date:
        yearList.append(eval(x.split('-')[0]))
    reviews['year'] = yearList
```

#### Merging reviews data with bussiness data

```python
# create user tag by merging
user_tag = data_review.merge(data_business, left_on='business_id', right_on='business_id')
user_tag.columns
```

```python
user_tag.shape
```

#### Generating User Tag

```python
dummy = user_tag.groupby('user_id').sum()
dummy.shape
```

```python
# creating a group by
user_tag2 = user_tag.groupby('user_id').sum().iloc[:,12:]
user_tag2.columns
```

```python
user_tag2.head(5)
```

```python
# creating tags
tag = []
for i in range(user_tag2.shape[0]):
    tmp = user_tag2.iloc[i,:].sort_values(ascending=False)[:2].index.tolist()
    tag.extend(tmp)

#len(set(tag))

tag_all = user_tag2.idxmax(axis=1)
tag_list = list(set(tag_all))
tag_all_df = pd.DataFrame(tag_all).reset_index()
```

```python
# updating tags

for tag in tag_list:
    data_users[tag] = 0
for ind,val in enumerate(tag_all):
    row = list(data_users['user_id']).index(tag_all_df['user_id'][ind])
    data_users.loc[row,val] = 1
    
data_users.iloc[:,24:].describe()

# creating outfile
```
