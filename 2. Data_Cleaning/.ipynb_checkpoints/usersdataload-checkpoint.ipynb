```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import json
```

```python
users= []
path = '/content/drive/MyDrive/CMPE256/Data/yelp_academic_dataset_user.json'
```

###Loading user data from Yelp Dataset

```python
with open(path, encoding = 'utf-8') as fin:
    i = 0 
    for line in fin:
        line_contents = json.loads(line)
        users.append(line_contents)
```

```python
data_user = pd.DataFrame(users)
data_user.shape
```

#### Writing coverted user data to CSV format

```python
data_user.to_csv("users_CA.csv")
```

```python
data_user.columns
```

#### Reading the California Specific User Reviews

```python
data_review = pd.read_csv('/content/drive/MyDrive/CMPE256/Data/CSVfiles/California_reviews_c.csv',index_col = 0)
data_review.head()
```

```python
data_review = data_review.drop(columns=['text'])
```

#### Extracting unique California user ids

```python
unique_CA_users = pd.DataFrame({'user_id': data_review.user_id.unique()})
unique_CA_users.head(5)
```

#### Merging CA unique user ids with Yelp user data

```python
d2 = data_user
intersected_df = d2.merge(unique_CA_users, how='left', on=['user_id'], indicator=True)
intersected_df.head(5)
```

```python
user_df = intersected_df.loc[intersected_df['_merge'] == 'both']
user_df.shape
```

```python
user_df = user_df.replace(r'^\s*$', np.nan, regex=True)
user_df
```

```python
user_df.isnull().sum()
```

#### Creating Is_elite column for users


```python
user_df['is_elite'] = ~np.array(user_df.elite.isnull())
user_df.is_elite.describe()
```

#### Adding User's Joining year in Yelp 

```python
# get the year
user_joining_year = [eval(x.split('-')[0]) for x in user_df.yelping_since]
user_df['year'] = user_joining_year
user_df = user_df.drop(columns=['_merge'])
```

#### Exporting final user data to CSV format

```python
user_df.to_csv("/content/drive/MyDrive/CMPE256/Data/CSVfiles/CA_users_c.csv")
```
